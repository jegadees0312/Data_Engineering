```
# Loading csv file
df = spark.read.csv("file:/Workspace/Users/azuser2115_mml.local@techademy.com/data.txt", header=True)
df.show()
```

**VISUALIZING USING MATPLOT**

```
import matplotlib.pyplot as plt

# convert to pandas
df_pandas = df.groupBy("CARRIER").agg({"ARR_DELAY":"avg"}) .toPandas()

# plot using MatplotLib
df_pandas.plot(kind = 'bar',x='CARRIER',y='avg(ARR_DELAY)',legend=True)
plt.title("Average Arrival Delay By Airline")
plt.xlabel("Airline")
plt.ylabel("Average Delay (minutes)")
plt.show()
```
